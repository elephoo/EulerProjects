```python
def pythagorean_triple(a,b,c):
        return a**2 + b**2 == c**2

triples = []
for a in range(1,500):
    for b in range(a + 1,500):
        for c in range(b + 1, 1000 - (a + b)):
            if pythagorean_triple(a,b,c):
                triples.append((a,b,c, a + b + c))
```

```python
# plot the triple as 3 2d plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
triples = pd.DataFrame(triples, columns=['a', 'b', 'c', 'sum'])
triples = triples.sort_values(by='sum')
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(triples['sum'], triples['a'])
plt.xlabel('sum')
plt.ylabel('a')
plt.subplot(3, 1, 2)
plt.plot(triples['sum'], triples['b'])
plt.xlabel('sum')
plt.ylabel('b')
plt.subplot(3, 1, 3)
plt.plot(triples['sum'], triples['c'])
plt.xlabel('sum')
plt.ylabel('c')
plt.tight_layout()
plt.show()
```
